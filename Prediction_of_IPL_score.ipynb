```python
df = df.drop(['date', 'runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5','mid', 'striker', 'non-striker'], axis =1)
```

```python
x = df.drop('total', axis = 1)
y = df['total']
x.info()
```

```python
from sklearn.preprocessing import LabelEncoder


venue_encoder = LabelEncoder()
batting_team_encoder = LabelEncoder()
bowling_team_encoder = LabelEncoder()
striker_encoder = LabelEncoder()
bowler_encoder = LabelEncoder()


venue_encoder.fit(x['venue'])
batting_team_encoder.fit(x['bat_team'])
bowling_team_encoder.fit(x['bowl_team'])
striker_encoder.fit(x['batsman'])
bowler_encoder.fit(x['bowler'])


x['venue'] = venue_encoder.transform(x['venue'])
x['bat_team'] = batting_team_encoder.transform(x['bat_team'])
x['bowl_team'] = bowling_team_encoder.transform(x['bowl_team'])
x['batsman'] = striker_encoder.transform(x['batsman'])
x['bowler'] = bowler_encoder.transform(x['bowler'])

x.info()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scale_std=StandardScaler()
x_train_std=scale_std.fit_transform(x_train)
x_test_std=scale_std.transform(x_test)
```

```python
!pip install keras-tuner
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import losses

def build_model(hp):
    model = Sequential()
    model.add(Dense(units=hp.Int('units1', min_value=64, max_value=256, step=64),
                    activation='relu', input_shape=(x_train.shape[1],)))
    model.add(Dense(units=hp.Int('units2', min_value=32, max_value=128, step=32), activation='relu'))
    model.add(Dense(1, activation='linear'))

    model.compile(
        optimizer=hp.Choice('optimizer', values=['adam', 'sgd']),
        loss=losses.Huber(delta=hp.Float('delta', min_value=0.5, max_value=1.5, step=0.1))
    )
    return model

tuner = kt.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=10,
    executions_per_trial=1,
    directory='my_dir',
    project_name='helloworld'
)

tuner.search(x_train_std, y_train, epochs=10, validation_split=0.2)

best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"Best Parameters: {best_hps.values}")
```

```python
model= tuner.get_best_models(num_models=1)[0]
model.summary()
```

```python
model.fit(x_train_std, y_train, epochs=10, validation_split=0.2)
```

```python
model.evaluate(x_test_std, y_test)
```

```python
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output


dropdown_style = {'description_width': 'initial'}
button_style = {'button_color': '#007BFF', 'font_weight': 'bold'}
output_style = {'font_size': '14px', 'border': '1px solid #ccc', 'padding': '10px', 'background': '#f9f9f9'}

venue = widgets.Dropdown(
    options=df['venue'].unique().tolist(),
    description='Select Venue:',
    style=dropdown_style,
    layout=widgets.Layout(width='400px')
)
batting_team = widgets.Dropdown(
    options=df['bat_team'].unique().tolist(),
    description='Select Batting Team:',
    style=dropdown_style,
    layout=widgets.Layout(width='400px')
)
bowling_team = widgets.Dropdown(
    options=df['bowl_team'].unique().tolist(),
    description='Select Bowling Team:',
    style=dropdown_style,
    layout=widgets.Layout(width='400px')
)
striker = widgets.Dropdown(
    options=df['batsman'].unique().tolist(),
    description='Select Striker:',
    style=dropdown_style,
    layout=widgets.Layout(width='400px')
)
bowler = widgets.Dropdown(
    options=df['bowler'].unique().tolist(),
    description='Select Bowler:',
    style=dropdown_style,
    layout=widgets.Layout(width='400px')
)

predict_button = widgets.Button(
    description="Predict Score",
    style=button_style,
    layout=widgets.Layout(width='200px', margin='10px 0')
)

output = widgets.Output(layout=widgets.Layout(width='100%', **output_style))


display(venue, batting_team, bowling_team, striker, bowler, predict_button, output)

def predict_score(b):
    with output:
        clear_output()

        decoded_venue = venue_encoder.transform([venue.value])
        decoded_batting_team = batting_team_encoder.transform([batting_team.value])
        decoded_bowling_team = bowling_team_encoder.transform([bowling_team.value])
        decoded_striker = striker_encoder.transform([striker.value])
        decoded_bowler = bowler_encoder.transform([bowler.value])


        input_data = np.array([decoded_venue, decoded_batting_team, decoded_bowling_team, decoded_striker, decoded_bowler])
        input_data = input_data.reshape(1, -1)
        input_data = scale_std.transform(input_data)


        predicted_score = model.predict(input_data)
        predicted_score = int(predicted_score[0, 0])

        print(f"Predicted Score: {predicted_score}")

predict_button.on_click(predict_score)
```
